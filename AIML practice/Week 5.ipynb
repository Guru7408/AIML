# Handle Outliers

# IQR method

```python
Q1 = df3['age'].quantile(0.25)
Q3 = df3['age'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR
```

```python
outliers = df3[(df3['age'] < lower) | (df3['age'] > upper)]
print("Univariate Outliers in Age:\n", outliers)
```

```python
# finding the 1st quartile
q1 = df["age"].quantile(0.25)

# finding the 3rd quartile
q3 = df['age'].quantile(0.75)

# finding the iqr region
iqr = q3-q1

# finding upper and lower whiskers
upper_bound = q3+(1.5*iqr)
lower_bound = q1-(1.5*iqr)
```

```python
upper_bound
```

```python
lower_bound
```

```python
age_arr = df["age"]
outliers = age_arr[(age_arr <= lower_bound) | (age_arr >= upper_bound)]
print(outliers)
```
